### Background
The election integrity partnership (which includes the Center for Informed Public at the University of Washington) released a [report](https://purl.stanford.edu/tr171zs0069) detailing how misinformation around the 2020 election spread. This included narratives around COVID-19 and how it complicated the political situation during the election. It also introduces what misinformation and disinformation are, and general ways in which they are spread. Misinformation is an umbrella term describing the spread of false, misleading or exaggerated information or claims. Disinformation is a subset of misinformation which describes the purposeful spread of misinformation with the intent to manipulate a specific narrative. 

A UWM report covers how researchers "found a direct correlation between locations where Twitter misinformation originated and subsequent spikes in COVID-19 infections and deaths in those areas weeks later." They also found that there were seven dominating myths around COVID-19 which was propogated in misinformed tweets in 2020. These myths centered around:

-Covid is over

-Economic liberty over safety

-Fake cures

-Flu comparison/ not serious

-Religious

-Ridiculing experts/ containment

-Individual liberty over safety


Knowing these popular narratives informed my research questions as I wanted to know if data from a different source would show similar narratives. More information can be found here: https://uwm.edu/news/mapping-the-harm-of-covid-19-misinformation-on-social-media/

This dataset was also used to inform a research paper by Samikshya Siwakoti, Kamya Yadav, Isra Thange, Nicola Bariletto, Luca Zanotti, Alaa Ghoneim, and Jacob N. Shapiro around "Localized Misinformation in a Global Pandemic: Report on COVID-19 Narratives around the World." ### Findings
Filtering out the COVID-19 misinformation stories not spread in the US left 441 distinct misinformation stories in the dataset. Out of ten possible COVID-19 misinformation narratives, the narrative that was centered in the most misinformation stories in the US was around government responses, with 105 distinct stories. According to the code book, a story coded as having a main narrative around government responses had to report "false information regarding government or political responses to fight the outbreak." This means that most misinformation stories in the US tried to falsify what actions the government was taking around COVID-19. This could be like declaring that a lockdown was ordered when it wasn't or vice versa. Notably, these stories must directly cite a government or governmental organization. For example, talking about the lockdown but not mentioning that it was a government sactioned lockdown would put the misinformation story in the other category. 

![](Narratives.png)

Out of 9 possible motives for spreading COVID-19 misinformation narratives, the most popular motive was politics with 141 stories. To qualify as a political motive, the narrative had to "target specific political actors or groups." More specifically, the story also had to aim to either "Discredit and weaken them" or "Help their political standing." 

![](Motives.png)

Out of the stories about only the most popular main narrative, government response, 56 out of 105 of those stories was motivated by politics. 

![](NarrativebyMotive.png)

These findings show that the disinformation narrative that was spread the most in 2020 was politically motivated to change opinions on the government's response.

As long as the story is distinct, it can be added to the dataset. This uncertainty around how the data is collected detracts from the credibility of my findings. 

Another limitation of this work is that the analysis of the data is slightly different from its original purpose. As stated before, the dataset was meant to be used to analyze global trends. By disregarding that and only analyzing stories from the US, some nuance and depth can be lost as the dataset was meant to look at the US as a whole and not on a regional level. 

### Conclusion
In 2020, disinformation around COVID-19 was a major issue, as little was known around COVID-19 at the start of the pandemic which occurred in the US around March, 2020. By analyzing what types of disinformation was being spread during this time period, the actions of different groups around this time can be contextualized. My analysis found that the most spread disinformation stories in the US in 2020 were around government responses to COVID-19. Further, the point of these stories was to motivate political change, whether it was to minimize the political power of the groups mentioned or to boost the political power of an author's own group. By understanding this, we recommend that the general public should be especially wary of COVID-19 stories around politics or the government, as they have a high possibility of being a disinformation story.

Import the [pandas library](https://pandas.pydata.org/docs/) to supplement summative analysis and cleaning the data. Import the [matplotlib.pyplot](https://matplotlib.org/stable/index.html) library in order to make graphs. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

Open the Covid misinformation excel spreadsheet and convert it into a dataframe to more easily manipulate it. Keep only the columns necessary for analysis i.e. main narrative, motive, country, and possibly publication date. Refine that to only stories from the United states.

```python
# read in data
covid_data = pd.read_excel(r"covid_misinfo.xlsx")

# change excel file to csv
covid_data.to_csv("covid_misinfo.csv", index=None, header=True)
df = pd.DataFrame(pd.read_csv("covid_misinfo.csv"))
print(df.keys())
print(len(df.keys()))

# exclude columns not necessary for analysis, keeping only: main narrative, motive, country, publication date
analysis_data = df[['Publication_Date', 'Primary_Country', 'Main_Narrative', 'Motive']]

# keep only stories from the United States
us_data = analysis_data[analysis_data['Primary_Country'] == 'United States']
```

Understand a brief overview of the dataset by getting a summary for the main narrative and motive columns, including how many data values there are, how many unique values there are, what the most frequent value is, and how often the most frequent value occurs.

```python
# find the most popular main narrative and motive for the disinformation stories in the dataset
print(us_data['Main_Narrative'].describe())
print(us_data['Motive'].describe())
```

Create a list of possible graph colors for future use.

```python
# graph references
New_Colors = ['green','blue','purple','brown','teal']
```

Create a bar chart showing the most popular disinformation narratives in the US in 2020. 

```python
# create a graph of the count of each main narrative
narratives = pd.unique(us_data['Main_Narrative'])
narrative_dict = {}
for value in narratives:
    narrative_dict[value] = 0

for row in us_data['Main_Narrative']:
    narrative_dict[row] = narrative_dict[row] + 1
    
print(narrative_dict)

narratives = [narrative.replace(' ', '\n') for narrative in narratives]
figure(figsize=(15, 12), dpi=80)
plt.bar(narratives, list(narrative_dict.values()), color=New_Colors)
plt.title('Main Narratives in Disinformation Stories by Count', fontsize=14)
plt.xlabel('Main Narrative', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.gcf()
plt.savefig('Narrative.png')
```

Create a bar chart for the most popular motives surrounding disinformation stories in the US in 2020.

```python
# create a graph of the count of each motive
motives = pd.unique(us_data['Motive'])
motive_dict = {}

for value in motives:
    motive_dict[value]=0

for row in us_data['Motive']:
    motive_dict[row]=motive_dict[row]+1

print(motive_dict)
motives = [motive.replace(' ', '\n') for motive in motives]
figure(figsize=(12, 12))
plt.bar(motives, list(motive_dict.values()), color=New_Colors)
plt.title('Motives in Disinformation Stories by Count', fontsize=14)
plt.xlabel('Motives', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.gcf()
plt.savefig('Motives.png')
```

Create a stack bar chart for only the most popular disinformation narrative and look at the break down for each type of motive.

```python
# create a stacked bar chart of the most popular narrative by motive
combined_data = us_data[us_data['Main_Narrative'] == 'Government responses']
combined_dict = {}
for key in motive_dict.keys():
    combined_dict[key]=0

for row in combined_data['Motive']:
    combined_dict[row]+=1
    
print(combined_dict)
    
last_value = 0
for value in combined_dict.values():
    plt.bar("Government Responses", value, bottom=last_value)
    last_value+=value
plt.xlabel("Main Narrative")
plt.ylabel("Motives")
plt.legend(motive_dict.keys(), bbox_to_anchor=(1, 0.5))
plt.title("Top Narrative by Motive")
plt.gcf()
plt.savefig('NarrativebyMotive.png')
```
